# Import Relevant Modules

```python
import os
import cv2
import numpy
from IPython.display import Image
from IPython.core.display import HTML 
```

# Set up a configuration

```python
config = {'height':700,
         'width':1200,
         'hue-lower':50,
         'hue-upper':120,
         'sat-lower':40,
         'sat-upper':255,
         'value-lower':140,
         'value-upper':255,
         'image-dir':'./data/',
         'area-threshhold':1000}
```

Code: https://stackoverflow.com/questions/47483951/how-to-define-a-threshold-value-to-detect-only-green-colour-objects-in-an-image/47483966#47483966

https://stackoverflow.com/questions/48109650/how-to-detect-two-different-colors-using-cv2-inrange-in-python-opencv

https://docs.opencv.org/3.1.0/dd/d49/tutorial_py_contour_features.html


```python
files = [config['image-dir'] + filename for filename in os.listdir(config['image-dir'])]
```

```python
files[:3]
```

```python
test_img_path = files[25]
```

```python
Image(filename = test_img_path, 
      width = config['width'], 
      height = config['height'])
```

```python
# Read in image
image = cv2.imread(test_img_path)

# Convert Image to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Create mask in HSV
mask = cv2.inRange(hsv, 
                   (config['hue-lower'], config['sat-lower'], config['value-lower']), 
                   (config['hue-upper'], config['sat-upper'], config['value-upper']))

# Slice the BLUE
imask = mask>0
pool_blue = numpy.zeros_like(image, numpy.uint8)
pool_blue[imask] = image[imask]

_, contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

#cv2.drawContours(pool_blue, contours, -1, (0, 255, 0), 1)

found_these = []
for c in contours:
    x,y, w, h = cv2.boundingRect(c)
    area = w * h
    if area >= config['area-threshhold']:
        found_these.append((x,y,w,h))
        cv2.rectangle(pool_blue, (x,y), (x+w,y+h), (0,255,0),2)

# Save result 
cv2.imwrite('test_result.png', pool_blue)

# Show 
Image(filename = './test_result.png', width = config['width'], height = config['height'])
```
